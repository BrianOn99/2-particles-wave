```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

m=1; n=2
X=np.linspace(0,1,30)
Y=np.linspace(0,1,30)
X,Y=np.meshgrid(X,Y)
phi_m = lambda x: sqrt(2)*np.sin(m*pi* x)
phi_n = lambda x: sqrt(2)*np.sin(n*pi* x)
Z_dist = phi_m(X)*phi_n(Y)
Z_sym  = (1/sqrt(2)) * (phi_m(X)*phi_n(Y)+phi_m(Y)*phi_n(X))
Z_anti = (1/sqrt(2)) * (phi_m(X)*phi_n(Y)-phi_m(Y)*phi_n(X))
```

```python
def plot3(X, Y, Z1, Z2, Z3, zlabel, title):
    fig = plt.figure(figsize=(20,5))
    ax1 = fig.add_subplot(131, projection='3d')
    ax2 = fig.add_subplot(132, projection='3d')
    ax3 = fig.add_subplot(133, projection='3d')
    axfig = ((ax1,Z1), (ax2,Z2), (ax3,Z3))
    for ax,Z in axfig:
        ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0.2, antialiased=True)
        #ax.view_init(elev=35, azim=30)
        ax.set_xlabel("$x_1/L$")
        ax.set_ylabel("$x_2/L$")
        ax.set_zlabel(zlabel)
 
    ax1.set_title("distinct")
    ax2.set_title("symmetric")
    ax3.set_title("anti-symmetric")

    fig.suptitle(title, fontsize=15)
    
plot3(X, Y, Z_dist, Z_sym, Z_anti, "$\psi$", "2 particles in a box -- Wave Function")
```

```python
plot3(X, Y, np.square(Z_dist), np.square(Z_sym), np.square(Z_anti),
      r"$|\psi|^2$", "2 particles in a box -- Prob. Distribution")
```
